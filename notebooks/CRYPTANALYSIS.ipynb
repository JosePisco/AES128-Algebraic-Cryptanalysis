# Cryptanalysis

## https://doc.sagemath.org/html/en/reference/cryptography/sage/crypto/mq/sr.html#sage.crypto.mq.sr.SR

Small Scale Variants of the AES (SR) Polynomial System Generator

Sage supports polynomial system generation for small scale 
(and full scale) AES variants over F2
and F2e. Also, Sage supports both the specification of SR 
as given in the papers [CMR2005] and [CMR2006] and a variant 
of SR* which is equivalent to AES.

SR is a family of parameterizable variants of the AES suitable 
as a framework for comparing different cryptanalytic techniques 
that can be brought to bear on the AES. It is different from Mini-AES,
whose purpose is as a teaching tool to help beginners 
understand the basic structure and working of the full AES.

## We construct SR(1,1,1,4) and study its properties.

```python
# n is the number of rounds, 
# r the number of rows in the state array, 
# c the number of columns in the state array, 
# e the degree of the underlying field.
n=2
r=2
c=2
e=4
sr = mq.SR(n, r,c, e)
```

```python
sr.n, sr.r, sr.c, sr.e
```

```python
# By default variables are ordered reverse to as they appear, e.g.:
print(sr.R.repr_long())
```

```python
# However, this can be prevented by passing in reverse_variables=False 
# to the constructor.
# For SR(1, 1, 1, 4) the ShiftRows matrix isn’t that interesting.:
sr.ShiftRows
```

```python
# Also, the MixColumns matrix is the identity matrix.:
sr.MixColumns
```

```python
# Lin, however, is not the identity matrix.:
sr.Lin
```

```python
# M and Mstar are identical for SR(1, 1, 1, 4):
```

```python
print(sr.M)
```

```python
print(sr.Mstar)
```

```python
# We can compute a Groebner basis for the ideals spanned by 
# SR instances to recover all solutions to the system.:
sr = mq.SR(1,1,1,4, gf2=True, polybori=True)
K = sr.base_ring()
a = K.gen()
K = [a]
P = [1]
F,s = sr.polynomial_system(P=P, K=K)
print("F: ",F)
print("s: ",s)
F.groebner_basis()
```

```python
# All solutions can easily be recovered using the variety function 
# for ideals.:
I = F.ideal()
for V in I.variety():
    for k,v in sorted(V.items()):
        print("{} {}".format(k, v))
    print("\n")
```

# Using these building blocks we can implement a simple XL algorithm  easily:

```python
sr = mq.SR(1,1,1,4, gf2=True, polybori=True, order='lex')
F,s = sr.polynomial_system()
```

```python
monomials = [a*b for a in F.variables() for b in F.variables() if a<b]
len(monomials)
```

```python
F2 = Sequence(map(mul, cartesian_product_iterator((monomials, F))))
A,v = F2.coefficient_matrix(sparse=False)
```

```python
A.echelonize()
A
```
