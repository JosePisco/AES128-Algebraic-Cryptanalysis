```python
def init_polynomial_sequence(n, r, c, e):
    sr = mq.SR(n, r, c, e, gf2=True, polybori=True, allow_zero_inversions=True)
    f, _ = sr.polynomial_system()
    return f.groebner_basis()
```

```python
def ideal_solving(f):
    i = f.ideal()
    for variety in i.variety():
        for k, v in sorted(variety.items()):
            print(k, v)
```

```python
poly_seq = init_polynomial_sequence(1, 1, 1, 4)
ideal_solving(poly_seq)
```
